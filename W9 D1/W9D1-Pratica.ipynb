```python
# Stringa iniziale
my_string = "I am studying Python"

# Stringa in maiuscolo (uppercase)
print("1• Maiuscolo:", my_string.upper())

# Stringa in minuscolo (lowercase)  
print("2• Minuscolo:", my_string.lower())

# Sostituzione "Python" con "a lot"
print("3• Sostituzione:", my_string.replace("Python", "a lot"))

# Usare .strip() non cambia nulla perché non ci sono spazi in testa ed in coda in my_string, con my_string2 cambia perché elimina gli spazi
print("4.1• Strip:", my_string.strip())
print("4.2• Cambia qualcosa? No.", my_string.strip())

my_string2= " I am studying python "
print("4.3• Cambia qualcosa? Ora si.", my_string2.strip())

# .split() per trovare tutte le parole
print("5• Parole (split):", my_string.split())
```
